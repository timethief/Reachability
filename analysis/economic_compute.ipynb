```python
economic_data = pd.read_csv('./data/ca_od_main_JT00_2010.csv')
```

```python
economic_data = economic_data[['w_geocode', 'S000', 'SA01', 'SA02', 'SA03', 'SE01', 'SE02', 'SE03']]
economic_data = economic_data.groupby('w_geocode').sum()
economic_data.reset_index(inplace = True)
```

```python
census_block_db = pd.read_csv('./data/net_am_peak/block_am_closest_node2.csv')
```

```python
economic_pd = economic_data[['w_geocode','S000','SE01','SE02','SE03']].merge(census_block_db[['GEOID10','node_id','net_lat','net_lon']], left_on='w_geocode', right_on='GEOID10')
```

```python
economic_pd = economic_pd[['w_geocode','S000','SE01','SE02', 'SE03', 'node_id']]
```

```python
economic_pd['S000'].sum(axis=0)
```

```python
store = pd.HDFStore('./data/net_late_night/net_late_night.h5', "r")
nodes = store.nodes
edges = store.edges
net=pdna.Network(nodes["x"], nodes["y"], edges["from"], edges["to"],
                 edges[["weight"]])
```

```python
def aggregrate():
    Distance = [20,40,60]
    print "Above"
    sys.stdout.flush()
    name = 'block_jobs_night_above'
    net.set(economic_pd['node_id'], variable=economic_pd.SE03.astype(float), name=name)
    start = time.time()
    for distance in Distance:
        print distance, " minutes"
        sys.stdout.flush()
        start = time.time()
        jobs_num = net.aggregate(distance, type='sum', decay='flat', name=name)
        end = time.time()
        print end - start, "s"
        sys.stdout.flush()
        jobs_pd = jobs_num.to_frame()
        jobs_pd.reset_index(inplace = True)
        jobs_pd = jobs_pd.merge(census_block_db[['node_id','net_lat','net_lon']], left_on='id_int', right_on='node_id',how='right')
        jobs_pd.to_csv('./data/net_late_night/jobs_night_above_'+ str(distance) + '.csv')
        
    print "All"
    sys.stdout.flush()
    name = 'block_jobs_night_all'
    net.set(economic_pd['node_id'], variable=economic_pd.S000.astype(float), name=name)
    start = time.time()
    for distance in Distance:
        print distance, " minutes"
        sys.stdout.flush()
        start = time.time()
        jobs_num = net.aggregate(distance, type='sum', decay='flat', name=name)
        end = time.time()
        print end - start, "s"
        sys.stdout.flush()
        jobs_pd = jobs_num.to_frame()
        jobs_pd.reset_index(inplace = True)
        jobs_pd = jobs_pd.merge(census_block_db[['node_id','net_lat','net_lon']], left_on='id_int', right_on='node_id',how='right')
        jobs_pd.to_csv('./data/net_late_night/jobs_night_all_'+ str(distance) + '.csv')
    
    print "Mediam"
    sys.stdout.flush()
    name = 'block_jobs_night_mediam'
    net.set(economic_pd['node_id'], variable=economic_pd.SE02.astype(float), name=name)
    start = time.time()
    for distance in Distance:
        print distance, " minutes"
        sys.stdout.flush()
        start = time.time()
        jobs_num = net.aggregate(distance, type='sum', decay='flat', name=name)
        end = time.time()
        print end - start, "s"
        sys.stdout.flush()
        jobs_pd = jobs_num.to_frame()
        jobs_pd.reset_index(inplace = True)
        jobs_pd = jobs_pd.merge(census_block_db[['node_id','net_lat','net_lon']], left_on='id_int', right_on='node_id',how='right')
        jobs_pd.to_csv('./data/net_late_night/jobs_night_mediam_'+ str(distance) + '.csv')
        
    print "Below"
    sys.stdout.flush()
    name = 'block_jobs_night_below'
    net.set(economic_pd['node_id'], variable=economic_pd.SE01.astype(float), name=name)
    start = time.time()
    for distance in Distance:
        print distance, " minutes"
        sys.stdout.flush()
        start = time.time()
        jobs_num = net.aggregate(distance, type='sum', decay='flat', name=name)
        end = time.time()
        print end - start, "s"
        sys.stdout.flush()
        jobs_pd = jobs_num.to_frame()
        jobs_pd.reset_index(inplace = True)
        jobs_pd = jobs_pd.merge(census_block_db[['node_id','net_lat','net_lon']], left_on='id_int', right_on='node_id',how='right')
        jobs_pd.to_csv('./data/net_late_night/jobs_night_below_'+ str(distance) + '.csv')
        
    print "all done"
        
```

```python
def build_heap_data(job_data):
    heat_lat = []
    heat_lon = []
    for i in range(0, len(job_data)):
        row = job_data.iloc[i]
        job_num = int(row[0] / 1000000)
        if row[0] > 0 and job_num == 0:
            job_num = 1
        for j  in range(0, job_num):
            heat_lat.append(row['net_lat'])
            heat_lon.append(row['net_lon'])
    return heat_lat, heat_lon
```

```python
heat_lat ,heat_lon = build_heap_data(jobs_pd)
```

```python
gmap = gmplot.GoogleMapPlotter.from_geocode('Los Angeles')
gmap.heatmap(heat_lat, heat_lon)       
#gmap.scatter(heat_data['net_lat'].values.tolist(),heat_data['net_lon'].values.tolist(),'r',marker=True)
gmap.draw("./data/job_map.html")
```
